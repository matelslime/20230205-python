```python
import time

#常用時間模組 posix, struct_time, str格式
# time.sleep(n) #暫停n秒
time.time() #local的posix時間 1970/01/01 00:00:00為起始 單位:秒 colab用的是google的local
#用來記錄經過時間
t1 = time.time()
l = []
for i in range(10**7):
  l.append(i)

t2 = time.time()
print(t2-t1)
```

```python
#非常用
now = time.time()
t = time.localtime(now)
print(t) #tm_mday月份的第幾日, tm_wday於星期一時顯示0, tm_yday年份的第幾日, tm_isdst日光節約時間
gm = time.gmtime(now) #格林威治時間
print("gm:",gm)
mk = time.mktime(gm) #時間格式反轉換 time.mktime(struct_time)
print("mk:",mk)

c = time.ctime(now) #C語言時間格式 (字串)
print("c:",c)
```

```python
import datetime as dt

d1 = dt.date(1799,2,4) #將時間格式化為date物件 可以處理1970以前時間
print(d1.year, d1.month, d1.day)
d2 = dt.date(year=2077,month=12,day=4)
#會產生datetime.timedelta物件
print(d2-d1)

t1 = dt.time(hour=10,minute=42,second=8,microsecond=15098)
print(t1.hour, t1.minute, t1.second, t1.microsecond)
t2 = dt.time(hour=22,minute=13,second=6,microsecond=532999)
#t2-t1 #不可相減

#dt.date(2021,2,29) #如不是閏年會超出範圍

dt.datetime(2022,3,4,11,3,56) #至少要輸入年月日 其他會預設0
dt.datetime.today()
dt.datetime.now() #datetime.today() = datetime.now() 都是現在
utc = dt.datetime.utcnow() #格林威治
print(utc.year, utc.day, utc.microsecond)

###直接用datetime就好
d = dt.datetime.date(1988, 10, 30)
dn = dt.datetime.now()
print(dn-d)

d.weekday()
d.isoweekday() #正常的weekday
d.isocalendar() #國際標準化的日曆格式 (正常的weekday)
d.isoformat(sep=" ") #國際標準化的時間字串格式 括號不須帶入東西 中間的T為分隔符 iso8601
d.ctime() #C語言的時間字串格式
d.strftime("") #自訂格式 可參考https://www.programiz.com/python-programming/datetime/strftime
d.strftime("%y") #只顯示年份後兩個
d.strftime("%Y") #完整年份
d.strftime("%b") #Jan
d.strftime("%B") #January
d.strftime("%m") #01
d.strftime("%-m") #1
d.strftime("%a") #Sun
d.strftime("%A") #Sunday
d.strftime("%w") #6
d.strftime("%d") #day of month
d.strftime("%H") #24小時制
d.strftime("%I") #12小時制
d.strftime("%p") #12小時制的am pm
d.strftime("%M") #月份
d.strftime("%S") #秒
d.strftime("%f") #microsecond
d.strftime("%j") #day of year
d.strftime("%W") #week of year
###strftime(), strptime() 最重要
print(d.strftime("%Y %B %a %-m %B %f")) #轉為字串
print(dt.datetime.strptime("01-09-2023","%m-%d-%Y")) #轉為時間物件 前為日期, 後為指定

dt.datetime.fromtimestamp(time.time()) #posix轉為字串

#timedelta()時間差物件
d_delta = dt.timedelta(weeks=3,days=5,hours=50,minutes=200,seconds=100) #將不同時間合計為幾天, 幾秒
print(dt.datetime.now()-d_delta) #可加減 年跟月因為會有閏年所以轉換為秒
```

##繼承

image.png

```python
class Car():
  def __init__(self):
    self.brand = "Toyota"
    self.doors = 4
    self.mileage = 0
  def run(self, x):
    self.mileage +=x
    print(x)

class Bus(Car):
  pass

bus1=Bus()
print(bus1.brand)
```

image.png

```python
class Bus(Car): # 繼承父類 Car
  def run(self, x): # 覆寫方法
    if self.mileage > 5000:
      print("進廠保養")
    else:
      self.mileage += x
  def greet(self): # 新增方法
    print("你好，歡迎搭乘")

bus2=Bus()
bus2.run(5001)
bus2.run(1)
bus2.greet()

car1=Car()
car1.run(4500)
```

```python
class Bus(Car): # 繼承父類 Car
  def __init__(self):
    super().__init__()
    self.person = "man"

  def greet(self): # 新增方法
    print("你好，歡迎搭乘")

bus4=Bus()
bus4.run(1)
bus4.doors #可以取得原類別
bus4.greet()
```

```python
class Animal:
  pass
class Bird (Animal):
  def fly (self):
    print("fly")
class Pigeon (Bird):
  pass
class Sparrow (Bird):
  pass

print(issubclass (Sparrow, Pigeon))
print(issubclass (Sparrow, Bird))
print(issubclass (Sparrow, Animal))
```

```python
#多重繼承
class Animal:
  def eat (self):
    print("animal eat")

class Bird:
  def eat (self):
    print ("bird eat")

class Duck (Animal, Bird):
  pass

duck=Duck()
duck.eat()
print(issubclass (Duck, Animal))
print(issubclass (Duck, Bird))
```
