```python
d1={"a":1,"b":2,"c":3}
d2={"d":4,"e":5,"f":6}

d1.update(d2)
print(d1)

d=d1.pop("d")
print(d)
print(d1)

kl=d1.keys()
print(kl)

vl=d1.values()
print(vl)

gl=d1.get("h",34)
print(d1)
print(gl)

sl=d1.setdefault("h",34)
print(d1)
print(sl)
```

```python
#5
def GCF(s1,s2):
  n1,n2=int(s1),int(s2)
  while True:
    if n2>n1:
      n1,n2=n2,n1

    if n1%n2==0 or n1%n2==1:
      return n2
    else:
      n1,n2=n2,n1%n2

print("最大公因數是:",GCF(input("請輸入一數"),input("請輸入一數")))
```
