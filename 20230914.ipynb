```python
def f(src):
  #open(src,'rb')讀二進位檔
  #open(src,'w')寫文字檔
  #open(src,'wb')寫二進位檔案
  #open(src,'w+')追加寫文件
  with open(src) as f: #默認為open(src,'r')
    data=f.read()

  key="Time,Ampl\n"
  data2=data[data.find(key)+len(key):-1]
  data3=data2.split("\n")

  Time=[];Ampl=[]
  for i in data3:
    a,b=i.split(",")
    Time.append(float(a)*int(b))
    Ampl.append(int(b))

  avg=sum(map(lambda x,y:x*y,Time,Ampl))/sum(Ampl)
  return avg

url1="data1.txt"
url2="data2.txt"
f(url1)
```

###範例

image.png

```python
#1
def fibonacci(n):
  if n==1:
    return 0
  if n==2:
    return 1
  else:
    return fibonacci(n-1)+fibonacci(n-2)

fibonacci(4)
```

```python
#2
def hanoi(n, A, B, C):
    if n == 1:
        print(f"{A}->{C}")
    else:
        hanoi(n-1, A, C, B)
        hanoi(1, A, B, C)
        hanoi(n-1, B, A, C)

n = int(input())
hanoi(int(n), 'A', 'B', 'C')
```

```python
def myDeco(fn):
  def wrapper(x):
    print("修改部分1")
    fn(x)
    print("修改部分2")
  return wrapper

@myDeco
def test(n):
  print("原本",n)

test(5)
```
