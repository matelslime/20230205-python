```python
import time

n = 20 #設定進度條總長
for i in range(n+1):
  # print(f'\r[{"❚"*i}{" "*(n-i)}]{i*100/n}%', end='') #輸出不換行的內容
  print('\r'+"❚"*i+" "*(n-i) + str(i*100/n)+'%', end='') # 輸出不換行的內容
  time.sleep(0.5)
```

```python
import time

n = 100
icon="⋮⋰⋯⋱"
# 建立旋轉的符號清單
for i in range(n+1):
  print(f'\r {icon[i%4]} {i*100/n}%', end='')
  time.sleep(0.2)
```
